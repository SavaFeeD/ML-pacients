```python
import os

import pandas as pd
import numpy as np
import collections

import matplotlib.pyplot as plt
import seaborn as sns

from sqlalchemy import create_engine

from joblib import Parallel, delayed
from tqdm import tqdm

from sklearn import preprocessing

import pickle
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
engine = create_engine('sqlite:///db.sqlite')
```

```python
y_cols = ['RecordID', 'Survival', 'In_hospital_death']
y = engine.execute("""
    select RecordID, Survival, In_hospital_death from hospital_stats
""").fetchall()
y[:10]
```

```python
X_cols = ['Weight', 'GCS', 'HR', 'NIDiasABP', 'NIMAP', 'NISysABP', 'RespRate',
       'Temp', 'Urine', 'HCT', 'BUN', 'Creatinine', 'Glucose', 'HCO3', 'Mg',
       'Platelets', 'K', 'Na', 'WBC', 'pH', 'PaCO2', 'PaO2', 'DiasABP', 'FiO2',
       'MAP', 'MechVent', 'SysABP', 'SaO2', 'Albumin', 'ALP', 'ALT', 'AST',
       'Bilirubin', 'Lactate', 'Cholesterol', 'TroponinI', 'TroponinT',
       'RecordID', 'Hour', 'Minute']
X = engine.execute("""
    select Weight, GCS, HR, NIDiasABP, NIMAP, NISysABP, RespRate,
       Temp, Urine, HCT, BUN, Creatinine, Glucose, HCO3, Mg,
       Platelets, K, Na, WBC, pH, PaCO2, PaO2, DiasABP, FiO2,
       MAP, MechVent, SysABP, SaO2, Albumin, ALP, ALT, AST,
       Bilirubin, Lactate, Cholesterol, TroponinI, TroponinT,
       RecordID, Hour, Minute from pacient_stats
""").fetchall()
X[:10]
```

```python
df = pd.DataFrame(X, columns=X_cols)
df.head()
```

```python
df_X = pd.DataFrame(X, columns=X_cols)
df_X.head()
```

```python
df_y = pd.DataFrame(y, columns=y_cols)
df_y.head()
```

```python
df['Survival'] = np.nan
df['In_hospital_death'] = np.nan
df.head()
```

```python
for idx, record_id in enumerate(tqdm(df['RecordID'].values)):
    df['Survival'].iloc[idx] = df_y[df_y['RecordID'] == record_id]['Survival']
    df['In_hospital_death'].iloc[idx] = df_y[df_y['RecordID'] == record_id]['In_hospital_death']
```

```python
df.head()
```

# Отбор признаков

```python
plt.figure(figsize=(55, 30))

matrix = np.triu(df.corr())
sns.heatmap(df.corr(), annot=True, mask=matrix)
```

С помощью корреляции Пирсона я выяснил что больше всего на выживаемость пациента в отделении интенсивной терапии  влияют показатели: HR, Lactate, Na, GCS, Time (Hour, Minute), BUN, Creatinine, RespRate, HCO3, DiasABP.

```python
cols = ['HR', 'Lactate', 'Na', 'GCS', 'Hour', 'Minute', 'Creatinine', 'RespRate', 'BUN', 'HCO3', 'DiasABP', 'In_hospital_death']
```

```python
new_df = pd.DataFrame(columns=cols)
new_df
```

```python
def processing(idx):
    return {col: df[col].iloc[idx] for col in cols}

new_df = new_df.append(Parallel(n_jobs=-1, verbose=1000)(delayed(processing)(idx) for idx in range(df.shape[0])))
```

```python
df = new_df

target_names = df['In_hospital_death'].unique().tolist()
target_names_new = ['Умер', 'Жив']

def replace_target(target):
    if target == target_names[0]:
        return target_names_new[0]
    if target == target_names[1]:
        return target_names_new[1]

df['In_hospital_death'] = df['In_hospital_death'].apply(lambda target: replace_target(target))   

target_names = df['In_hospital_death'].unique().tolist()

df.head()
```

# Разбиение набора данных на выборки

Разбил выборки в соотношении 6:2:2 (обучающая:валидационная:тестовая) потому что на валидационную и тестирующую выборки нет необходимости выделять больше данных из данного датасета для данной задачи

```python
y_name = 'In_hospital_death'
```

```python
train, validate, test = np.split(df.sample(frac=1), [int(.6*len(df)), int(.8*len(df))])
sns.barplot(data=pd.DataFrame([{'train': train.shape[0], 'validate': validate.shape[0], 'test': test.shape[0]}]))
```

# Выбор модели

Целевая переменная, содержит информацию умер человек в отделение интенсивной терапии (1) или нет (0)

Для такой задачи необходимо использовать алгорим классификации

Модели для выбора:
1. RandomForestClassifier
2. KNeighborsClassifier
3. CatBoostClassifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from catboost import CatBoostClassifier
```

### Настройка гипер параметров

Пометил гипер параметры и примерный разброс их значений и с помощью HalvingRandomSearchCV подобрал нужные гипер параметры для моделей

```python
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingRandomSearchCV

from sklearn.metrics import classification_report
```

```python
X_validate = validate.drop(y_name, axis=1)
y_validate = validate[y_name]
```

```python
models = {
    'rf': RandomForestClassifier(),
    'kn': KNeighborsClassifier(),
    'cb': CatBoostClassifier()
}
```

```python
hyper_params = {
    'rf': {
        'max_depth': range(1, 10),
        'n_jobs': [-1],
        'criterion': ['gini', 'entropy']
    },
    'kn': {
        'algorithm': ['ball_tree', 'kd_tree', 'brute'],
        'n_neighbors': range(1, 10),
        'weights': ['uniform', 'distance'],
        'n_jobs': [-1]
    },
    'cb': {
        'iterations': range(500, 2000, 500),
        'silent': [True]
    }
}
```

```python
best_params = {}

for model_name in tqdm(models):
    search = HalvingRandomSearchCV(models[model_name],
                                   hyper_params[model_name],
                                   random_state=0).fit(X_validate, y_validate)
    best_params[model_name] = search.best_params_
```

```python
best_params
```

## Тестирование моделей

Для тестирования и выбора модели буду использовать метрики F1-score, precision, recall, accurasy.

```python
X_train = train.drop(y_name, axis=1)
y_train = train[y_name]

X_test = test.drop(y_name, axis=1)
y_test = test[y_name]
```

```python
models = {
    'rf': RandomForestClassifier(**best_params['rf']),
    'kn': KNeighborsClassifier(**best_params['kn']),
    'cb': CatBoostClassifier(**best_params['cb'])
}
```

```python
metrics = {
    target_names[0]: [],
    target_names[1]: []
}

for model_name in tqdm(models):
    y_pred = models[model_name].fit(X_train, y_train).predict(X_test)
    report = classification_report(y_test, y_pred, target_names=target_names, output_dict=True)
    
    m1 = report[target_names[0]]
    m1['model'] = model_name
    
    m2 = report[target_names[1]]
    m2['model'] = model_name
    
    metrics[target_names[0]].append(m1)
    metrics[target_names[1]].append(m2)    
```

```python
df_metrics_0 = pd.DataFrame(metrics[target_names[0]])
df_metrics_0 = df_metrics_0.drop('support', axis=1)
print(target_names[0])
df_metrics_0.sort_values(['precision', 'recall', 'f1-score'])
```

```python
df_metrics_1 = pd.DataFrame(metrics[target_names[1]])
df_metrics_1 = df_metrics_1.drop('support', axis=1)
print(target_names[1])
df_metrics_1.sort_values(['precision', 'recall', 'f1-score'])
```

Отсортировав по всем показателям и вывев таблицы с метриками по классам можно увидеть как определяют разные алгоритмы все классы

Лучше всего пределяет класс 'Умер' алгоритм KNeighborsClassifier, а класс 'Жив' алгоритм RandomForestClassifier поэтому посмотрим метрику accuracy (точность), чтобы решить какой алгоритм справляется лучше с поставленной задачей

```python
from sklearn.metrics import accuracy_score
```

```python
metric_accuracy = {}

for model_name in tqdm(models):
    y_pred = models[model_name].fit(X_train, y_train).predict(X_test)
    metric_accuracy[model_name] = accuracy_score(y_test, y_pred)
```

```python
pd.DataFrame([{key: f'{str(metric_accuracy[key]*100)[:4]}%' for key in metric_accuracy.keys()}])
```

У алгоритма RandomForestClassifier точность выше чем у KNeighborsClassifier поэтому я выбираю алгоритм RandomForestClassifier для задачи классификации умер человек в отделение интенсивной терапии или нет
