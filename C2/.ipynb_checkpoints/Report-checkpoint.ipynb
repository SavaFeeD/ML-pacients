```python
import os

import pandas as pd
import numpy as np
import collections

import matplotlib.pyplot as plt
import seaborn as sns

from sqlalchemy import create_engine

from joblib import Parallel, delayed
from tqdm import tqdm

from sklearn import preprocessing

import pickle
```

```python
import warnings
warnings.filterwarnings('ignore')
```

# Сбор данных из базы

```python
engine = create_engine('sqlite:///db.sqlite')
```

```python
y_cols = ['RecordID', 'Survival', 'In_hospital_death']
y = engine.execute("""
    select RecordID, Survival, In_hospital_death from hospital_stats
""").fetchall()
y[:10]
```

```python
X_cols = ['Weight', 'GCS', 'HR', 'NIDiasABP', 'NIMAP', 'NISysABP', 'RespRate',
       'Temp', 'Urine', 'HCT', 'BUN', 'Creatinine', 'Glucose', 'HCO3', 'Mg',
       'Platelets', 'K', 'Na', 'WBC', 'pH', 'PaCO2', 'PaO2', 'DiasABP', 'FiO2',
       'MAP', 'MechVent', 'SysABP', 'SaO2', 'Albumin', 'ALP', 'ALT', 'AST',
       'Bilirubin', 'Lactate', 'Cholesterol', 'TroponinI', 'TroponinT',
       'RecordID', 'Hour', 'Minute']
X = engine.execute("""
    select Weight, GCS, HR, NIDiasABP, NIMAP, NISysABP, RespRate,
       Temp, Urine, HCT, BUN, Creatinine, Glucose, HCO3, Mg,
       Platelets, K, Na, WBC, pH, PaCO2, PaO2, DiasABP, FiO2,
       MAP, MechVent, SysABP, SaO2, Albumin, ALP, ALT, AST,
       Bilirubin, Lactate, Cholesterol, TroponinI, TroponinT,
       RecordID, Hour, Minute from pacient_stats
""").fetchall()
X[:10]
```

```python
df = pd.DataFrame(X, columns=X_cols)
df.head()
```

```python
df_X = pd.DataFrame(X, columns=X_cols)
df_X.head()
```

```python
df_y = pd.DataFrame(y, columns=y_cols)
df_y.head()
```

```python
df['Survival'] = np.nan
df['In_hospital_death'] = np.nan
df.head()
```

```python
for idx, record_id in enumerate(tqdm(df['RecordID'].values)):
    df['Survival'].iloc[idx] = df_y[df_y['RecordID'] == record_id]['Survival']
    df['In_hospital_death'].iloc[idx] = df_y[df_y['RecordID'] == record_id]['In_hospital_death']
```

Датасет готов

```python
df.head()
```

# Отбор признаков

```python
plt.figure(figsize=(55, 30))

matrix = np.triu(df.corr())
sns.heatmap(df.corr(), annot=True, mask=matrix)
```

С помощью корреляции Пирсона я выяснил что больше всего на выживаемость пациента в отделении интенсивной терапии  влияют показатели: HR, Lactate, Na, GCS, Time (Hour, Minute), Creatinine, RespRate.

```python
cols = ['HR', 'Lactate', 'Na', 'GCS', 'Hour', 'Minute', 'Creatinine', 'RespRate', 'Survival']
```

```python
new_df = pd.DataFrame(columns=cols)
new_df
```

```python
def processing(idx):
    return {col: df[col].iloc[idx] for col in cols}

new_df = new_df.append(Parallel(n_jobs=-1, verbose=1000)(delayed(processing)(idx) for idx in range(df.shape[0])))
```

```python
new_df.head()
```

```python
df = new_df
```

# Разбиение набора данных на выборки

Разбил выборки в соотношении 6:2:2 (обучающая:валидационная:тестовая) потому что на валидационную и тестирующую выборки нет необходимости выделять больше данных из данного датасета для данной задачи

```python
train, validate, test = np.split(df.sample(frac=1), [int(.6*len(df)), int(.8*len(df))])
sns.barplot(data=pd.DataFrame([{'train': train.shape[0], 'validate': validate.shape[0], 'test': test.shape[0]}]))
```

# Выбор модели

Модели для выбора:
1. LinearRegression
2. SVR
3. CatBoostRegressor

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from catboost import CatBoostRegressor
```

```python
X_validate = validate.drop('Survival', axis=1)
y_validate = validate['Survival']

scaler = preprocessing.MinMaxScaler()
d = scaler.fit_transform(X_validate)

scaled_X_validate = pd.DataFrame(d, columns=X_validate.columns)
scaled_X_validate.head()
```

### Настройка гипер параметров

Пометил гипер параметры и примерный разброс их значений и с помощью HalvingRandomSearchCV подобрал нужные гипер параметры для моделей

```python
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingRandomSearchCV

from sklearn.model_selection import cross_val_score
```

```python
models = {
    'lr': LinearRegression(),
    'svr': SVR(),
    'cb': CatBoostRegressor()
}
```

```python
hyper_params = {
    'lr': {
        'fit_intercept': [True, False],
    },
    'svr': {
        'C': np.linspace(1.0, 3.0),
        'epsilon': np.linspace(0.1, 0.5)
    },
    'cb': {
        'iterations': [500, 1000, 2000],
        'silent': [True]
    }
}
```

```python
best_params = {}

for model_name in tqdm(models):
    search = HalvingRandomSearchCV(models[model_name],
                                   hyper_params[model_name],
                                   max_resources=10,
                                   random_state=0).fit(scaled_X_validate, y_validate)
    best_params[model_name] = search.best_params_
```

```python
best_params
```

## Тестирование моделей

Для тестирования и выбора модели буду использовать метрику кросс валидация 

```python
X_test = test.drop('Survival', axis=1)
y_test = test['Survival']

scaler = preprocessing.MinMaxScaler()
d = scaler.fit_transform(X_test)

scaled_X_test = pd.DataFrame(d, columns=X_test.columns)
scaled_X_test.head()
```

```python
models = {
    'lr': LinearRegression(best_params['lr']),
    'svr': SVR(**best_params['svr']),
    'cb': CatBoostRegressor(**best_params['cb'])
}
```

```python
cross_score = {}

for model_name in tqdm(models):
    cross_score[model_name] = cross_val_score(models[model_name], scaled_X_test, y_test, cv=10)
```

```python
cross_score
```

```python
plt.figure(figsize=(20, 7))
sns.lineplot(data=cross_score)
```

У модели CatBoostRegressor показатели метрики кросс валидации выше чем у других моделей поэтому я выбираю ее для прогноза продолжительности жизни пациента

# Сохранение модели

```python
X_train = test.drop('Survival', axis=1)
y_train = test['Survival']

scaler = preprocessing.MinMaxScaler()
d = scaler.fit_transform(X_train)

scaled_X_train = pd.DataFrame(d, columns=X_train.columns)
scaled_X_train.head()
```

```python
model = CatBoostRegressor(**best_params['cb']).fit(scaled_X_train, y_train)
```

```python
with open('model_r.pkl', 'wb') as f:
    pickle.dump(model, f)
```
